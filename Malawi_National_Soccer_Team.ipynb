```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv(r'C:\Users\victo\Downloads\My Dataset\Dataset_Malawi_National_Football_Team_Matches.csv')

# Count match outcomes
result_counts = df['Result'].value_counts()

# Print counts
#print("Match outcomes:")
#print(result_counts)

# Plotting
plt.figure(figsize=(15, 6))
result_counts.plot(kind='bar', color=['green', 'red', 'gray'])
plt.title('Malawi National Team Match Outcomes')
plt.ylabel('Number of Games')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
df.columns

# Group by Opponent and calculate average or total opponent score
opponent_scores = df.groupby('Opponent')['Opponent Score'].mean()

# Plot
plt.figure(figsize=(15, 6))
opponent_scores.plot(kind='bar', color='blue')
plt.title('Average Opponent Score vs Malawi by Opponent')
plt.ylabel('Average Score')
plt.xlabel('Opponent')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Filter for games Malawi lost
losses = df[df['Result'] == 'Loss']

# Group by Competition and Venue, then count losses
grouped = losses.groupby(['Competition', 'Venue']).size().unstack(fill_value=0)

# Plot as a grouped bar chart
grouped.plot(kind='bar', figsize=(15, 6))
plt.title('Malawi Losses by Competition and Venue')
plt.ylabel('Number of Losses')
plt.xlabel('Competition')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Count how many times Malawi played each opponent
opponent_counts = df['Opponent'].value_counts()

# Plot
plt.figure(figsize=(12, 6))
opponent_counts.plot(kind='bar', color='steelblue')
plt.title('Number of Matches Malawi Played Against Each Opponent')
plt.xlabel('Opponent')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Malawi vs Mozambique Match Results
# Filter for matches against Mozambique
mozambique_matches = df[df['Opponent'] == 'Mozambique']

# Count results (Win, Loss, Draw)
result_counts = mozambique_matches['Result'].value_counts()

# Plot
plt.figure(figsize=(6, 5))
result_counts.plot(kind='bar',figsize=(15, 5), color=['blue', 'red', 'green'])
plt.title('Malawi vs Mozambique Match Results')
plt.xlabel('Result')
plt.ylabel('Number of Matches')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

```
